# Visualizing Data

## What is matplotlib?
Matplotlib is the most popular and mature library for plotting data using Python. It has all of the functionality you would expect, including the ability to control the formatting of plots and figures at a very fine level.

### Importing Matplotlib

```python
#Just as we use the np shorthand for NumPy and the pd shorthand for Pandas.

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### First Plot

```python
plt.plot(np.random.normal(size=200), np.random.normal(size=200), 'o')
```

### Plotting from an IPython notebook

```python
%matplotlib inline
```

```python
#After running this command any cell within the notebook that creates a plot will embed a PNG image of the resulting graphic.

plt.plot(np.random.normal(size=200), np.random.normal(size=200), 'd')
```

## Plotting in Pandas

```python
#Pandas includes methods for DataFrame and Series objects that are relatively high-level.

enter = pd.Series(np.random.normal(size=45))
print(enter)
enter.plot()
```

```python
#By default a line plot is drawn, and a light grid is included.

enter.cumsum().plot(grid=True)
```

```python
#Similarly, for a DataFrame.

datas = pd.DataFrame({'normal': np.random.normal(size=15), 
                       'gamma': np.random.gamma(1, size=15), 
                       'poisson': np.random.poisson(size=15)})
print(datas)

datas.cumsum(0).plot()
```

```python
#We can split multiple series into subplots with single argument per plot.

datas.cumsum(0).plot(subplots=True)
```

```python
#We may want to have some series displayed on the secondary y-axis, which can allow for greater detail and less empty space.

datas.cumsum(0).plot(secondary_y='normal')
```

```python
#We can use matplotlib's subplots function directly, and manually assign plots to its axes.

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 12))
for i,v in enumerate(['normal','gamma','poisson']):
    datas[v].cumsum(0).plot(ax=axes[i], title=v)
axes[0].set_ylabel('cumulative sum')
```

## Simple Line Plots

```python
#Perhaps the simplest of all plots is the visualization of a single function  =() . 

img = plt.figure()
axs = plt.axes()

s = np.linspace(0, 15, 98)
print(s)
axs.plot(s, np.sin(s));
```

```python
# We can use the pylab interface and let the figure and axes be created for us in the background.

plt.plot(s, np.sin(s))
```

## Line Colors and Styles

```python
#The plt.plot() function takes additional arguments that can be used to specify these. 

plt.plot(s, np.sin(s - 2), color='blue')        
plt.plot(s, np.sin(s - 1), color='g')           
plt.plot(s, np.sin(s - 0), color='0.55')       
```

```python
#The line style can be adjusted using the linestyle keyword.

plt.plot(s, np.sin(s - 0), linestyle='-')        
plt.plot(s, np.sin(s - 1), linestyle=':')           
plt.plot(s, np.sin(s - 2), linestyle='-.')  
```

## Adjusting the Plot: Axes Limits

```python
#The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods.

plt.plot(s, np.sin(s - 0))

plt.xlim(-1, 14)
plt.ylim(-2, 3);
```

## Labeling Plots

```python
#Titles and axis labels are the simplest such labelsthere are methods that can be used to quickly set them.

plt.plot(s, np.cos(s))
plt.title("Cosine Curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

## Simple Scatter Plots

```python
#The points are represented individually with a dot, circle, or other shape. 

plt.style.use('seaborn-whitegrid')
```

### Scatter Plots with plt.plot

```python
#These character codes can be used together with line and color codes to plot points along with a line connecting them.

x = np.linspace(1, 12, 65)
y = np.log(x)

plt.plot(x, y, 'o-', color='blue')
```

```python
plt.plot(x, y, '-og');
```

### Scatter Plots with plt.scatter

```python
#The plt.scatter function, which can be used very similarly to the plt.plot function.

plt.scatter(x, y, marker='o', color='red')
```

```python
#Use the alpha keyword to adjust the transparency level.

r = np.random.RandomState(7)
x = r.randn(120)
y = r.randn(120)
colors = r.rand(120)
sizes = 1000 * r.rand(120)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Reds')
plt.colorbar()  # show color scale
```

## Errorbars

```python
#A basic errorbar can be created with a single Matplotlib function call.

x = np.linspace(0, 10, 111)
r = 0.1
y = np.cos(x) + r * np.random.randn(111)

plt.errorbar(x, y, yerr=r, fmt='.y');
```

## Histograms

```python
#A simple histogram can be a great first step in understanding a dataset.

plt.style.use('seaborn-white')

omg = np.random.randn(9999)
print(omg)
```

```python
plt.hist(omg)
```

```python
#The hist() function has many options to tune both the calculation and the display.

plt.hist(omg, bins=5, normed=True, alpha=0.6,
         histtype='stepfilled')
```

```python
#Combination of histtype='stepfilled' along with some transparency alpha to be very useful when comparing histograms of several distributions.

x = np.random.normal(0, 0.8, 1000)
y = np.random.normal(-2, 1, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.5, normed=True, bins=50)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

## Two-Dimensional Histograms

```python
#Histograms in two-dimensions by dividing points among two-dimensional bins.

mean = [6, 3]
cov = [[4, 2], [3, 1]]
x, y = np.random.multivariate_normal(mean, cov, 9999)
```

### plt.hist2d: Two-dimensional histogram

```python
#One straightforward way to plot a two-dimensional histogram is to use Matplotlib's plt.hist2d function.

plt.hist2d(x, y, bins=25, cmap='Blues')
cb = plt.colorbar()
```

### plt.hexbin: Hexagonal binnings

```python
#Matplotlib provides the plt.hexbin routine, which will represents a two-dimensional dataset binned within a grid of hexagons.

plt.hexbin(x, y, gridsize=10, cmap='Greens')

cb = plt.colorbar()
```
